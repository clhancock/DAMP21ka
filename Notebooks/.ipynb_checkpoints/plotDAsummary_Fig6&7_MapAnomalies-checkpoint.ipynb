```python
#Select results to plot ('' will load the most recent assimilation file)
filename = ''
filename=filename.replace('/',':')

import os
wd = '/Users/chrishancock/Library/CloudStorage/OneDrive-NorthernArizonaUniversity/Research/Manuscript/DAMP21k/Github/DAMP21ka/' #changed
os.chdir(wd+'Holocene-code') #

# Math Functions
import numpy as np
import scipy.stats
import pandas as pd
import xarray as xr
# Plotting Functions
import matplotlib.pyplot   as plt         # Packages for making figures
import matplotlib.cm       as cm
import matplotlib.gridspec as gridspec
import cartopy.crs         as ccrs        # Packages for mapping in python
import cartopy.feature     as cfeature
import cartopy.util        as cutil
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
# DA Functions
import da_utils_plotting as da_plot
import da_load_proxies
import da_load_models
import da_psms
print('packages loaded')
font = {'family': 'sans-serif', 'sans-serif': 'Lucida Grande'}
```

**LOAD DATA**

```python
# If don't define a reconstruction, Get the most recent filename
if filename == '':
    filename = [f for f in os.listdir(wd+'Data/results') if os.path.isdir(os.path.join(wd+'Data/results', f))]
    filename = filename[np.argmax([os.path.getmtime(wd+'Data/results/'+f+'/'+f+'.nc') for f in filename])]
print('loading '+filename)
# Load Data
options, DAMPvals, DAMPproxy = da_plot.loadDAMPresults(wd+'Data/results/',filename,kalman=True)

#Define dataframe
proxy_df = pd.DataFrame(DAMPproxy['proxy_metadata'][:,[0,1,2,3,4,5,8,10,12]], columns=['dataSetName','TSid','lat','lon','seasonality_array','season','units','archive','PSM'])
#Convert lat/lons to float
proxy_df['lat'] = [np.float(x) for x in proxy_df['lat']];
proxy_df['lon'] = [np.float(x) for x in proxy_df['lon']];

#Define mapping settings
#options['assimilate_selected_region'] = [15.0, 65.0, 190.0, 360.0]
if options['assimilate_selected_region']:
    lims,proj = [], ccrs.PlateCarree()
    for x in options['assimilate_selected_region']:
        if x < 180: lims.append(x)
        else: lims.append(x-360)
    lims = np.array(lims)[[2,3,0,1]]
else:     
    lims,proj = False, ccrs.Robinson()
```

**MAP GlOBAL ANOMALIES CORRELATIONS**

```python
# Select which ka year (slect nearest to age*1000) to plot relative to the Last Millenium mean
times = [21,6]
variables = ['LakeStatus','precip'] 
labels = [['a','b','c'],['d','e','f']]
#For each type
for t in times:
    #Set up plot
    plt.figure(figsize=(6.5,6),dpi=400)
    plt.rc('font', **font)
    gs = gridspec.GridSpec(3,len(variables))
    #
    for col,var in enumerate(variables):
        #Load data for specific variable
        prior,recon = DAMPvals[var]['prior'].mean(dim='iteration'), DAMPvals[var]['recon'].mean(dim='iteration')
        proxies,proxies_psm = DAMPproxy['proxy_values'], DAMPproxy['proxyprior_mean'].mean(dim='iteration') 
        units=DAMPvals[var]['units'][0]
        if t == 21: t_i = np.where((prior['ages'].data<=t*1000)&(prior['ages'].data>=t*1000-1000))[0]
        else: t_i = np.where((prior['ages'].data<=t*1000+500)&(prior['ages'].data>=t*1000-500))[0]
        #Set as anomaly relative to last millenium mean
        prior -= prior.sel(ages=slice(0,1000)).mean(dim='ages')
        recon -= recon.sel(ages=slice(0,1000)).mean(dim='ages')
        proxies -= DAMPproxy['proxy_values'].sel(ages=slice(0,1000)).mean(dim='ages')
        proxies_psm -= DAMPproxy['proxyprior_mean'].sel(ages=slice(0,1000)).mean(dim=['iteration','ages'])
        #Calculate innovation
        innov = recon-prior 
        #Standardize colors
        if var == 'LakeStatus': v = 70
        elif var == 'tas': v=3.5
        else: v=0.7 #v = np.nanquantile(np.abs(recon),0.9) 
        if (t > 12) & (var != 'LakeStatus'): v*=2
        levs = np.linspace(-v,v,22)
        cmap = cm.get_cmap('BrBG')
        if var == 'tas': cmap=cm.get_cmap('coolwarm')
        #Plot each of prior recon and innovation
        for row,name in enumerate(['recon','prior','innov']):
            #Set up subplot
            ax = plt.subplot(gs[row,col],projection=proj)
            da_plot.plotBaseMap(ax,proj,lims)
            #Get specific data to plot
            if name == 'recon':
                vals_t, text =  recon[t_i],'reconstruction'
                proxyi = proxy_df[(proxy_df.PSM == 'get_'+var)]
                proxyvals=proxies[t_i,proxyi.index]
            else:
                proxyi = proxy_df[(proxy_df.PSM == 'get_'+var)]
                proxyvals = proxies[t_i,proxyi.index]
                if name == 'prior':   vals_t,text = prior[t_i],'model prior'
                elif name == 'innov': vals_t,text = innov[t_i],'update'; proxyvals -= proxies_psm[t_i,proxyi.index]
            #Calc timeslice mean
            if len(t_i)>1: proxyvals,vals_t = proxyvals.mean(axis=0),vals_t.mean(axis=0)
            #Plot variable values
            ax.set_title('('+labels[col][row]+')',fontsize=8,loc='left')
            if var == 'LakeStatus': ax.set_title(text+' - lake status',fontsize=8)
            elif var == 'precip':      ax.set_title(text+' - precipitation',fontsize=8)
            elif var == 'tas':      ax.set_title(text+' - temperature',fontsize=8)
            data_cyclic,lon_cyclic = cutil.add_cyclic_point(vals_t,coord=vals_t.lon.data)
            idx=np.isfinite(proxyvals.data)
            #idx=np.argsort(np.abs(proxyvals.data))[np.isfinite(proxyvals.data)]
            p=ax.contourf(lon_cyclic,vals_t.lat,data_cyclic,extend='both',levels=levs,cmap=cmap,transform=ccrs.PlateCarree())
            #Plot proxy values/sites
            if var == 'LakeStatus': plt.scatter(proxyi.lon[idx],proxyi.lat[idx],c=proxyvals[idx], cmap=cmap,vmin=-v,vmax=v,ec='k',linewidth=0.3,s=5,alpha=1,transform=ccrs.PlateCarree())
            else: plt.scatter(proxy_df.lon,proxy_df.lat,c='k',s=0.7,linewidth=0,transform=ccrs.PlateCarree())
        #Colorbar
        cbaxes = inset_axes(ax,width="80%",height="10%", loc="lower center", bbox_to_anchor=(0, -0.2, 1, 1),bbox_transform=ax.transAxes)
        cbar = plt.colorbar(p, extend='both',orientation='horizontal', ticks=list(np.round(np.linspace(-v,v,8),2)),cax=cbaxes)
        cbar.ax.tick_params(labelsize=8) 
        cbar.set_label(label='('+units+')',size=8)
    #Format
    plt.suptitle(str(t)+'-0 ka',fontsize=10,y=0.97,fontweight='bold')
    plt.tight_layout()
    #Save
    plt.savefig(wd+'Data/results/'+filename+'/'+'Fig6&7_'+'UpdateMap_'+var+'_'+str(t)+'ka.png', dpi=600, bbox_inches ="tight")

```
